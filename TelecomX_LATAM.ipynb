```python
df_flat = pd.json_normalize(
    df_raw.to_dict(orient="records")
)
```

```python
df_flat.head()
df_flat.info()
```

```python
df_flat.columns = (
    df_flat.columns
    .str.lower()
    .str.replace(".", "_")
)
```

```python
string_cols = df_flat.select_dtypes(include="object").columns

df_flat[string_cols] = df_flat[string_cols].apply(lambda x: x.str.lower())
```

```python
internet_services = [
    "internet_onlinesecurity",
    "internet_onlinebackup",
    "internet_deviceprotection",
    "internet_techsupport",
    "internet_streamingtv",
    "internet_streamingmovies"
]

for col in internet_services:
    df_flat[col] = df_flat[col].replace("no internet service", "no")
```

```python
df_flat["phone_multiplelines"] = df_flat["phone_multiplelines"].replace(
    "no phone service", "no"
)
```

```python
df_flat["account_charges_total"] = (
    pd.to_numeric(df_flat["account_charges_total"], errors="coerce")
)
```

```python
df_flat["account_charges_monthly"] = (
    pd.to_numeric(df_flat["account_charges_monthly"], errors="coerce")
)
```

```python
df_flat["customer_tenure"] = (
    pd.to_numeric(df_flat["customer_tenure"], errors="coerce")
)
```

```python
df_flat.info()
```

```python
df_flat.isnull().sum().sort_values(ascending=False)
```

```python
df_flat["cuentas_diarias"] = df_flat["account_charges_monthly"] / 30
```

```python
df_flat[["account_charges_monthly", "cuentas_diarias"]].head()
```

```python
df_flat["cuentas_diarias"].describe()
```

```python
binary_cols = [
    "churn",
    "customer_partner",
    "customer_dependents",
    "phone_phoneservice",
    "paperlessbilling",
    "internet_onlinesecurity",
    "internet_onlinebackup",
    "internet_deviceprotection",
    "internet_techsupport",
    "internet_streamingtv",
    "internet_streamingmovies"
]
```

```python
binary_map = {"yes": 1, "no": 0}

for col in binary_cols:
    if col in df_flat.columns:
        df_flat[col] = df_flat[col].map(binary_map)
```

```python
df_flat["customer_seniorcitizen"] = df_flat["customer_seniorcitizen"].astype(int)
```

```python
df_flat["account_contract"] = df_flat["account_contract"].replace({
    "month-to-month": "mensual",
    "one year": "anual",
    "two year": "bianual"
})
```

```python
df_flat["account_paymentmethod"] = df_flat["account_paymentmethod"].replace({
    "electronic check": "cheque_electronico",
    "mailed check": "cheque_fisico",
    "bank transfer (automatic)": "transferencia_bancaria",
    "credit card (automatic)": "tarjeta_credito"
})
```

```python
df_flat = df_flat.rename(columns={
    "customer_gender": "genero",
    "customer_tenure": "antiguedad_meses",
    "account_charges_monthly": "cargo_mensual",
    "account_charges_total": "cargo_total",
    "cuentas_diarias": "cargo_diario"
})
```

```python
import matplotlib.pyplot as plt

def churn_by_category(df, column):
    churn_rate = (
        df.groupby(column)["churn"]
        .mean()
        .sort_values(ascending=False)
    )

    plt.figure()
    churn_rate.plot(kind="bar")
    plt.title(f"Tasa de Churn segÃºn {column}")
    plt.xlabel(column)
    plt.ylabel("ProporciÃ³n de Churn")
    plt.show()
```

#**ðŸ“Š Informe Final â€“ AnÃ¡lisis de EvasiÃ³n de Clientes (Churn)**

**Challenge Telecom X**

##**1ï¸âƒ£ IntroducciÃ³n**

El objetivo de este anÃ¡lisis es comprender el fenÃ³meno de evasiÃ³n de clientes (churn) en la empresa Telecom X, identificando patrones y factores que influyen en la decisiÃ³n de los clientes de cancelar o no el servicio.

La evasiÃ³n de clientes representa un desafÃ­o estratÃ©gico para las empresas de telecomunicaciones, ya que implica pÃ©rdida directa de ingresos y mayores costos de adquisiciÃ³n de nuevos clientes. A travÃ©s de un proceso de ETL (ExtracciÃ³n, TransformaciÃ³n y Carga) y un AnÃ¡lisis Exploratorio de Datos (EDA), se busca obtener insights que permitan orientar acciones para reducir el churn y mejorar la retenciÃ³n de clientes.

##**2ï¸âƒ£ Limpieza y Tratamiento de Datos**
**ExtracciÃ³n de datos**

Los datos fueron obtenidos directamente desde la API de Telecom X en formato JSON. Este archivo contenÃ­a informaciÃ³n demogrÃ¡fica de los clientes, servicios contratados, detalles de facturaciÃ³n y estado de evasiÃ³n.

**TransformaciÃ³n y limpieza**

Durante esta etapa se realizaron las siguientes acciones:

- Aplanado de la estructura JSON utilizando json_normalize, ya que los datos se
encontraban anidados.

- NormalizaciÃ³n de nombres de columnas (minÃºsculas y uso de guiones bajos).

- ConversiÃ³n de variables numÃ©ricas que estaban almacenadas como texto (por ejemplo, cargos mensuales y totales).

- Tratamiento de valores inconsistentes en variables categÃ³ricas, como "no internet service" y "no phone service".

- ConversiÃ³n de variables binarias de texto (yes / no) a valores numÃ©ricos (1 / 0).

- CreaciÃ³n de una nueva variable cargo_diario, calculada a partir del cargo mensual, para obtener una visiÃ³n mÃ¡s detallada del comportamiento de gasto.

Este proceso permitiÃ³ obtener un dataset limpio, coherente y preparado para el anÃ¡lisis exploratorio.

##**3ï¸âƒ£ AnÃ¡lisis Exploratorio de Datos (EDA)**
**DistribuciÃ³n del churn**

El anÃ¡lisis inicial mostrÃ³ que la mayorÃ­a de los clientes permanece en la empresa, aunque existe un porcentaje significativo de clientes que cancelan el servicio, lo que justifica un anÃ¡lisis mÃ¡s profundo de los factores asociados a la evasiÃ³n.

**Churn y variables categÃ³ricas**

El anÃ¡lisis por variables categÃ³ricas revelÃ³ patrones relevantes:

- **Tipo de contrato:** los contratos mensuales presentan una tasa de churn considerablemente mÃ¡s alta que los contratos anuales o bianuales.

- **MÃ©todo de pago:** los clientes que utilizan ciertos mÃ©todos de pago, como cheque electrÃ³nico, muestran una mayor propensiÃ³n a cancelar.

- **Servicios adicionales:** la ausencia de servicios como soporte tÃ©cnico o seguridad en lÃ­nea se asocia con mayores tasas de evasiÃ³n.

- **FacturaciÃ³n sin papel:** se observan diferencias en el churn segÃºn la preferencia de facturaciÃ³n.

**Churn y variables numÃ©ricas**

El anÃ¡lisis de variables numÃ©ricas permitiÃ³ identificar diferencias claras entre clientes que cancelan y los que no:

- Los clientes que hacen churn suelen tener menor antigÃ¼edad en la empresa.

- Presentan cargos mensuales mÃ¡s elevados, lo que puede influir en la decisiÃ³n de cancelar.

- El gasto total acumulado es mayor en clientes que permanecen, explicado por su mayor tiempo de permanencia.

Las visualizaciones (boxplots e histogramas) reforzaron estos hallazgos, mostrando distribuciones claramente diferenciadas entre ambos grupos.

##**4ï¸âƒ£ Conclusiones e Insights**

A partir del anÃ¡lisis realizado, se destacan los siguientes hallazgos clave:

- La antigÃ¼edad del cliente es uno de los factores mÃ¡s influyentes en la evasiÃ³n.

- Los contratos de corto plazo presentan un riesgo significativamente mayor de churn.

- Un cargo mensual elevado, especialmente en clientes con baja antigÃ¼edad, incrementa la probabilidad de cancelaciÃ³n.

- La contrataciÃ³n de servicios adicionales estÃ¡ asociada a una mayor retenciÃ³n de clientes.

- Existen perfiles de clientes con mayor propensiÃ³n a churn que pueden ser identificados tempranamente.

Estos insights permiten comprender mejor el comportamiento de los clientes y sirven como base para acciones estratÃ©gicas.

##**5ï¸âƒ£ Recomendaciones**

Con base en los resultados del anÃ¡lisis, se proponen las siguientes recomendaciones:

- Incentivar contratos de mayor duraciÃ³n, ofreciendo beneficios o descuentos a clientes con contratos mensuales.

- Implementar estrategias de retenciÃ³n temprana para clientes nuevos, especialmente durante los primeros meses de servicio.

- Revisar la estructura de cargos mensuales elevados, ofreciendo planes mÃ¡s flexibles o personalizados.

- Promover la contrataciÃ³n de servicios adicionales, como soporte tÃ©cnico o seguridad en lÃ­nea, como elementos de fidelizaciÃ³n.

- DiseÃ±ar campaÃ±as especÃ­ficas para los perfiles de clientes con mayor riesgo de evasiÃ³n, basadas en los patrones identificados.

Este anÃ¡lisis demuestra cÃ³mo el uso de tÃ©cnicas de Data Science y AnÃ¡lisis Exploratorio de Datos permite transformar datos crudos en informaciÃ³n valiosa para la toma de decisiones.
